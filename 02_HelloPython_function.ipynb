```python
def mySum(a, b):
    return a+b
```

```python
c = mySum(2, 3)
print(c)
```

```python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
        
    def hello(self):
        print("Hello " + self.name + " !")
        
    def goodbye(self):
        print("Good-bye " + self.name + " !")
```

```python
m = Man("David")
```

```python
m.hello()
```

```python
m.goodbye()
```
